```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_pos=0
y_pos=0
x_direct=-1
y_direct=1
fig,ax=plt.subplots(figsize=(12,7))
ax.quiver(x_pos,y_pos,x_direct,y_direct,scale=10)
ax.set_title('quiver plot with one arrow')
plt.show()
```

```python
x_pos=[0,0]
y_pos=[0,0]
x_direct=[1,0]
y_direct=[-1,1]
fig,ax=plt.subplots(figsize=(12,7))
ax.quiver(x_pos,y_pos,x_direct,y_direct,color='red',scale=5)
ax.set_title('Quiver plot with two arrow')
plt.show()
```

```python
fig,ax1 = plt.subplots(figsize =(14, 8))

x = np.arange(0, 2.2, 0.2)
y = np.arange(0, 2.2, 0.2)
X, Y = np.meshgrid(x, y)
u = np.cos(X)*Y
v = np.sin(y)*Y
n = -2

color = np.sqrt(((v-n)/2)*2 + ((u-n)/2)*2)

ax1.quiver(X, Y, u, v, color, alpha = 0.8)
ax1.xaxis.set_ticks([])
ax1.yaxis.set_ticks([])
ax1.axis([-0.2, 2.3, -0.2, 2.3])
ax1.set_aspect('equal')
ax1.set_title('meshgrid function')
```

```python
x = np.arange(-1, 1, 0.1)
y = np.arange(-1, 1, 0.1)
X, Y = np.meshgrid(x, y)
u = np.cos(X)*Y
v = np.sin(y)*Y

fig, ax = plt.subplots(figsize=(12, 7))
ax.quiver(X, Y, u, v, color='b', alpha=0.5)

ax.set_title('Streamlined Wind Directions')
plt.show()
```

```python
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
u = np.sin(X*2 + Y*2)
v = np.cos(X*2 - Y*2)

plt.figure()
plt.quiver(X, Y, u, v)
plt.title('Quiver Plot with Arrows')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize =(14, 8))

x = np.arange(0, 2.2, 0.2)
y = np.arange(0, 2.2, 0.2)
X, Y = np.meshgrid(x, y)
u = np.cos(X)*Y
v = np.sin(y)*Y
n = -2

color = np.sqrt(((v-n)/2)*2 + ((u-n)/2)*2)

ax1.quiver(X, Y, u, v, color, alpha = 0.8)
ax1.xaxis.set_ticks([])
ax1.yaxis.set_ticks([])
ax1.axis([-0.2, 2.3, -0.2, 2.3])
ax1.set_aspect('equal')
ax1.set_title('meshgrid function')

x = np.arange(-2, 2.2, 0.2)
y = np.arange(-2, 2.2, 0.2)
X, Y = np.meshgrid(x, y)
z = X * np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)
n = -2


color = np.sqrt(((dx-n)/2)*2 + ((dy-n)/2)*2)


ax2.quiver(X, Y, dx, dy, color)
ax2.xaxis.set_ticks([])
ax2.yaxis.set_ticks([])
ax2.set_aspect('equal')
ax2.set_title('gradient')



plt.tight_layout()
plt.show()
```

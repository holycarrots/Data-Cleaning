```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Electric_Production.csv")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

#KDE Plot
df.plot(kind='kde', ax=axs[0, 1])
axs[0, 1].set_title('KDE Plot')

#Line Plot
df.plot(x='DATE', y='Value', kind='line', ax=axs[1, 0])
axs[1, 0].set_title('Line Plot')

#Another Line Plot in a different format
df['Value'].plot(kind='line', ax=axs[1, 1], figsize=(8, 4), title='Value')
axs[1, 1].set_title('Line Plot')

plt.tight_layout()

plt.show()
```

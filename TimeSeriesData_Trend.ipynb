```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data={"Date": pd.to_datetime(["2022-01-01",
              "2022-01-02",
              "2022-01-03",
              "2022-01-04",
              "2022-01-05",
              "2022-01-06",
              "2022-01-07",
              "2022-01-08",
              "2022-01-09",
              "2022-01-10",
              "2022-01-11",
              "2022-01-12",
              "2022-01-13",
              "2022-01-14",
              "2022-01-15"]),
      "Value":[10,15,22,28,35,42,48,55,60,68,75,82,90,98,105]}
```

```python
df = pd.DataFrame(Data)
```

```python
df.set_index("Date",inplace=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.index, df["Value"], marker=0)
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time series with trend")
plt.grid(True)
plt.show()
```
